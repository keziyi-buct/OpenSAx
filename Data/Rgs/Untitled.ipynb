```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 读取CSV文件
df= pd.read_csv('LUCAS.SOIL_corr.csv')
# 创建一个画布
fig, ax = plt.subplots(1, 3, figsize=(12, 4))

# 创建数据
l=np.arange(400,2499.5,0.5)

# 画第一条直线
data = df.values[:5000,:-9]
ax[0].plot(l,data.T)
ax[0].set_xlabel('Wavelength(nm)')
ax[0].set_ylabel('Reflectance')
ax[0].set_title('Small-scale dataset')
# 画第二条直线
data = df.values[:10000,:-9]
ax[1].plot(l,data.T)
ax[1].set_xlabel('Wavelength(nm)')
ax[1].set_ylabel('Reflectance')
ax[1].set_title('medium-scale dataset')

# 画第三条直线
data = df.values[:15000,:-9]
ax[2].plot(l,data.T)
ax[2].set_xlabel('Wavelength(nm)')
ax[2].set_ylabel('Reflectance')
ax[2].set_title('large-scale dataset')

plt.savefig("data_draw.jpg", dpi=300)
# 展示图形
plt.show()

```
